```python
%%manim -qm a

class a(Scene):
    def construct(self):
        plane = NumberPlane(
            x_range = (0, 25,2),
            y_range =(-11, 12,2),
            x_length = 10,
            y_length = 5,
            x_axis_config={"include_numbers": True},
        )
        plane.center()

        line_graph = plane.plot_line_graph(
            x_values = [i for i in range(25)],
            y_values = np.concatenate(([0], np.cumsum([np.random.choice([-1,1]) for i in range(25)]))),
            line_color=GOLD_E,
            vertex_dot_style=dict(stroke_width=3,  fill_color=PURPLE),
            stroke_width = 4,
        )
        self.add(plane, line_graph)
```
