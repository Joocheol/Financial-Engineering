```python
%%manim -qm a

N = 30
class a(Scene):
    def construct(self):
        plane = NumberPlane(
            x_range = (0, N, 5),
            y_range =(-11, 12,2),
            x_length = 10,
            y_length = 5,
            x_axis_config={"include_numbers": True},
        )
        plane.center()

        line_graph = plane.plot_line_graph(
            x_values = [0,1,1,2,3,3,5],
            y_values = [0,0,2,2,2,3,3],
            line_color=GOLD_E,
            vertex_dot_style=dict(stroke_width=3,  fill_color=PURPLE),
            stroke_width = 4,
        )
        self.add(plane)
        self.play(Create(line_graph['line_graph']), run_time=5)
        self.play(Create(line_graph['vertex_dots']), run_time=5)
        self.wait()
```

```python
%%manim -qm a

N = 30
class a(Scene):
    def construct(self):
        plane = NumberPlane(
            x_range = (0, N, 5),
            y_range =(-11, 12,2),
            x_length = 10,
            y_length = 5,
            x_axis_config={"include_numbers": True},
        )
        plane.center()

        line_graph = plane.plot_line_graph(
            x_values = [i for i in range(N)],
            y_values = np.concatenate(([0], np.cumsum([np.random.choice([-1,1]) for i in range(N)]))),
            line_color=GOLD_E,
            vertex_dot_style=dict(stroke_width=3,  fill_color=PURPLE),
            stroke_width = 4,
        )
        self.add(plane)
        self.play(Create(line_graph['line_graph']), run_time=5)
        self.play(Create(line_graph['vertex_dots']), run_time=5)
        self.wait()
```
